```python
from bs4 import BeautifulSoup
import pandas as pd 
import requests
import numpy as np
```

```python
def get_title(new_soup):
    try:
        title_string = new_soup.find("span", attrs={"id":'productTitle'}).text.strip()
    except AttributeError:
        title_string = ""
    return title_string

def get_auth(new_soup):
    try:
        auth_string = new_soup.find("span", attrs={"class":'author notFaded'}).text.strip()
        author = auth_string[:auth_string.index('\n')].strip()
    except AttributeError:
        author = ""
    return author

def get_lang(new_soup):
    try:
        language = new_soup.find("div", attrs= {"id":'rpi-attribute-language'}).text.strip()
        lang_string = language[9:].strip()
    except AttributeError:
        lang_string = ""
    return lang_string

def get_publisher(new_soup):
    try:
        pub = new_soup.find("div", attrs={"id":'rpi-attribute-book_details-publisher'}).text.strip()
        pub_string = pub[9:].strip()
    except AttributeError:
        pub_string = ""
    return pub_string

def get_rating(new_soup):
    try:
        rating = new_soup.find("span", attrs={"class":'a-icon-alt'}).text.strip()
        # rat_string = rating[:rating.index("out")]
        if rating == 'Previous page':
            rating = 'Not Available'
    except AttributeError:
        rating = ""
    return rating


```

```python
if __name__ == '__main__':
    URL = "https://www.amazon.in/s?k=new+books&i=stripbooks&rh=n%3A976389031%2Cn%3A1318168031&s=review-rank&dc&ds=v1%3A29N7kDSDZTMegHsJ8qkT8bJcVvpKj5vJF0JfxhKIlLQ&crid=29A5CXXO4VDPF&qid=1679839873&rnid=3576079031&sprefix=%2Caps%2C286&ref=sr_st_review-rank"
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.69','Accept-Language': 'en-US,en;q=0.5'})
    webpage = requests.get(URL, headers = HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    links= soup.find_all("a",attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    
    links_list = []
    for link in links:
       links_list.append(link.get('href'))
    
    image= soup.find_all("img",attrs={"class":'s-image'})
    image
    img_list = []
    for i in image:
        img_list.append(i.get('src'))
   

    d = {"title": [], "author": [], "publisher":[], "rating": [], "language": [], "image_link": []}

    for link in links_list:
        new_webpage = requests.get('https://amazon.com' + link, headers = HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, 'html.parser')

        d['title'].append(get_title(new_soup))
        d['author'].append(get_auth(new_soup))
        d['publisher'].append(get_publisher(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['language'].append(get_lang(new_soup))
        d['image_link'].append(img_list)

    sem6_df_rom = pd.DataFrame.from_dict(d)
    sem6_df_rom['title'].replace('',np.nan,inplace=True)
    sem6_df_rom = sem6_df_rom.dropna(subset=['title'])
    sem6_df_rom.to_csv("sem6_project_rom.csv", header=True, index = False)
        

    
```
